# TA-Lib Sandbox
Used to verify pattern recognition logic on historical stock candles.

```python
import yfinance as yf
import talib
import pandas as pd

# 1. Fetch Sample Data
ticker = 'HDFCBANK.NS'
df = yf.download(ticker, period='1mo', interval='1d')
df.head()

# 2. Run Test Pattern (Engulfing)
df['CDLENGULFING'] = talib.CDLENGULFING(df['Open'], df['High'], df['Low'], df['Close'])
print("Detected Engulfing Patterns:")
print(df[df['CDLENGULFING'] != 0])
```
